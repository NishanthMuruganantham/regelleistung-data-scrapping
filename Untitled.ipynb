```python
from lxml import html
import pandas as pd
import requests
from lxml.html import parse, submit_form, tostring
```

```python
page = requests.post("https://www.regelleistung.net/ext/data/?from=14.10.2021&_download=on&tsoId=4&dataType=MRL&lang=en")
data = html.fromstring(page.content)
```

```python
tso_xpath = "//select[contains(@name,'tsoId')]/option"

tso_option_list = data.xpath(tso_xpath)

available_tso = []

for tso in tso_option_list:
    available_tso.append(tso.get("value"))

print(available_tso)
```

```python
datatype_xpath = "//select[contains(@name,'dataType')]/option"

datatypes_option_list = data.xpath(datatype_xpath)

available_datatypes = []

for datatype in datatypes_option_list:
    available_datatypes.append(datatype.get("value"))

print(available_datatypes)
```

```python
def find_table_data(tso,data_type):
    page = requests.post(f"https://www.regelleistung.net/ext/data/?from=14.10.2021&_download=off&tsoId={tso}&dataType={data_type}&lang=en")
    tree = page.text #html.fromstring(page.content)
    tree = tree.replace("<div>","")
    tree = tree.replace("</div>","")
    try:
        df = pd.read_html(tree)
        return df
    except ValueError:
        return False
    
```

```python
for tso in available_tso:
    for data_type in available_datatypes:
        df = find_table_data(tso,data_type)
        print(df)
```
